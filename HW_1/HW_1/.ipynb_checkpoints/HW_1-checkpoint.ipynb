### Предобработка даты

Т.к. в данных очевидно есть какая-то сезонность, к примеру зимой температура явно уменьшается, признак даты является важным. Для удобства разобьем дату на 3 столбца: год, месяц, день.

```python
df['Date'] = pd.to_datetime(df.Date)
df['Year'] = df.Date.dt.year
df['Month'] = df.Date.dt.month
df['Day'] = df.Date.dt.day
df.drop(columns=['Date'], axis=1, inplace=True)
df
```

### Предобработка категориальных признаков

```python
cat_cols = ['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday']
```

#### Обработка пропущенных значений

```python
df[cat_cols].isna().sum() / df.shape[0]
```

Количество пропущенных  значений в категориальных признаков составляет менее 8%, т.к. это относительно мало можно просто удалить эти строки, посмотрим как сократится DS если удалить строки с пропущенными элементами:

```python
df.dropna()
```

Видно, что останется менее 50%, что неприемлемо. Исходя из того, что пропущенных значений в категориальных признаков мало, а в вещественных много, предлагается разделить DS на 2: в одном строки с пропущенными значениями в категориальных признаках удалятся, а а другом пропущенные значения заменятся 'NA',но т.к. по заданию размер датасета изменять нельзя, выберем только второй вариант

```python
df_cat = df[cat_cols].fillna('NA').astype(str)
df_cat.isna().any()
```

### Преобразование категориальных признаков

Многие алгоритмы ML не умеют работать с категориальными признаками, я лично знаю только один который умеет с ними работать - CatBoost. Рассмотрим наши категориальные признаки, задачу и доступные модели для выбора соответствующего методо преобразования.

```python
df[cat_cols]
```

Как видно все столбцы не являются порядковыми. Рассотрим "мощность" или как принято называть cardinality признаков.

```python
for i in cat_cols:
    print("{} cardinality: {}".format(i, df[i].unique().size))
    print("{} distribution: {}".format(i, df[i].value_counts()))
```

Для методов не являющимися деревьями я предпочитаю использовать one-hot encoding для задач классификации и похожий на него dummies encoding для задач регрессии, но у них есть проблема, т.к. количество признаков возрастает на n и n-1, где n-cardinality. Т.к. признак Location иммет высокую cardinality = 47, для него воспользуемся LeaveOneOut encoding. Для остальных признаков cadinality <= 17, что довольно много, но терпимо, для них воспользуемся one-hot encoding.

```python
from category_encoders.one_hot import OneHotEncoder
OHE_encoder = OneHotEncoder()
X_cat_no_loc = OHE_encoder.fit_transform(df_cat.drop(columns=['Location']))
X_cat_no_loc.shape
```

```python
from category_encoders.leave_one_out import LeaveOneOutEncoder
LOOE_encoder = LeaveOneOutEncoder()
X_cat_loc = LOOE_encoder.fit_transform(df_cat['Location'], target)
X_cat_loc.shape
```

```python
import numpy as np
X_cat = pd.concat([X_cat_loc, X_cat_no_loc], axis=1)
X_cat.shape
```

## Предобработка вещественных признаков

```python
numeric_cols = list(set(df.columns.values.tolist()) - set(cat_cols))
numeric_cols
```

Пропуски в вещественных признаках будем заменять нулями, т.к. как правило это дает лучший показатель по сравнению с ост. к примеру медианным или средним значением

```python
X_real = df[numeric_cols].fillna(0)
```

Посмотрим на данные

```python
X_real.describe()
```

Исходя из данных видно что столбцы: WindSpeed3pm, RISK_MM, Evaporation, WindGustSpeed, WindSpeed9am, Rainfall содержат большое количество выбросов, построим boxplot, чтобы в этом убедиться

```python
df.boxplot(column=['WindSpeed3pm', 'RISK_MM', 'Evaporation', 'WindGustSpeed', 'WindSpeed9am', 'Rainfall'])
```

Действительно, из графиков видно, что данные сталбцы имют огромное число выбросов, также видно, что столбцы Rainfall и RISK_MM - возможно коррелируют между собой. Для удаления этих выбросов, я бы воспользовался IQR-score, но строки в этом DS по заданию удалять нельзя(

Также видно, что разные признаки очень сильно отличаются друг от друга по модулю значений. В случае обычной регрессии это никак не влияет на качество обучаемой модели, т.к. у меньших по модулю признаков будут большие веса, но при использовании регуляризации, которая штрафует модель за большие веса, регрессия, как правило, будет работать хуже.
Проведем масштабирование признаков

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_real_scaled = scaler.fit_transform(X_real)
```

```python
X_scaled_cat = X_cat
X_scaled_cat[numeric_cols] = X_real_scaled
X_scaled_cat.describe()
```

Т.к. мы уже заметили коррелированные признаки, построим матрицу корреляций и выделим признаки, которые имеют коеф корреляции более 0.9

```python
corr_matrix = X_scaled_cat.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]
to_drop
```

```python
X_scaled_cat.drop(columns=to_drop, axis=1, inplace=True)
```

# Построение моделей

Посмотрим на сбалансированность классов:

```python
target.value_counts()
```

Видно, что классы несбалансированные. Т.к. изменять размер DS нельзя, а также нельзя использовать стратификацию, то будем использовать возможности моделей в sklearn. Также в качестве метрики будем использовать площать по PR кривой.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import average_precision_score
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled_cat, target, test_size=0.25, random_state=123, shuffle=False)
```

## Байесовский классификатор

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
```

```python
prediction_NB = clf.predict(X_test)
print("PR auc: ", average_precision_score(y_test, prediction_NB))
```

Попробуем улучшить качество, отобрав признаки используя p-value

```python
import statsmodels.api as sm
X_1 = sm.add_constant(X_train)
model = sm.OLS(y_train,X_1).fit()
model.pvalues.max()
```

Предлагаю удалять по одному признаку с макссимальным p-value, пока max(p-value) не будет меньше 0.05, и посмотреть как будет изменяться оценка модели:

```python
def get_result(X_train, X_test, y_train = y_train, y_test=y_test):
    clf = GaussianNB()
    clf.fit(X_train, y_train)
    prediction_NB = clf.predict(X_test)
    return average_precision_score(y_test, prediction_NB)

X_train_NB = X_train.copy()
X_test_NB = X_test.copy()
results = []
columns_drop = []
for i in range(X_scaled_cat.shape[1]):
    results.append(get_result(X_train_NB, X_test_NB))
    X_1 = sm.add_constant(X_train_NB)
    model = sm.OLS(y_train,X_1).fit()
    if(model.pvalues.max()>0.05):
        X_train_NB.drop(columns = model.pvalues.idxmax(), axis=1, inplace=True)
        X_test_NB.drop(columns = model.pvalues.idxmax(), axis=1, inplace=True)
        columns_drop.append(model.pvalues.idxmax())
        print(i)
    else:
        break
```

```python
plt.plot(columns_drop, results[:-1])
```

```python
max_index = results.index(max(results)) + 1
columns_drop = columns_drop[:max_index]
columns_drop
```

В итоге получаем следующий результат:

```python
%%time
clf = GaussianNB()
clf.fit(X_train_NB, y_train)
prediction_NB = clf.predict(X_test_NB)
print("PR auc: ", average_precision_score(y_test, prediction_NB))
```

## Логистическая регрессия

Выбирать оптимальные параметры для Логистической регресси будем с помощью поиска по сетке, также установим балансировку классов и будем использовать кросс валидацию на 3 фолда

```python
%%time
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.01, 0.05, 0.1, 0.5, 1, 5, 10],
              'penalty': ['l2']
             }
LR = GridSearchCV(LogisticRegression(class_weight='balanced'), param_grid, cv=3, scoring='average_precision')
LR.fit(X_train, y_train)
prediction_LR = LR.predict(X_test)
print("PR auc: ", average_precision_score(y_test, prediction_LR))
```

Что-то странное: результат качества максимальный, давайте посмотрим остальные метрики

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction_LR))
```

Возможно, что один из признаков слишком сильно коррелирует с таргетом, давайте посмотрим

```python
importance = LR.best_estimator_.coef_[0]
for i,v in enumerate(importance):
    print('Feature: {}, Score: {}'.format(X_train.columns[i],v))
plt.bar([x for x in range(len(importance))], importance)
plt.show()
```

Как видно RISK_MM, слишком сильно коррелирует с target, да, на kaggle в описании DS, так и написано, что ее нужно удалить). Для чистоты эксперимента удалим эту переменную

```python
X_train.drop(columns='RISK_MM', axis=1, inplace=True)
X_test.drop(columns='RISK_MM', axis=1, inplace=True)
```

Повторим наш эксперимент:

```python
%%time
LR.fit(X_train, y_train)
prediction_LR = LR.predict(X_test)
print("PR auc: ", average_precision_score(y_test, prediction_LR))
```

Результат не очень хороший, посмотрим на остальные метрики:

```python
print(classification_report(y_test, prediction_LR))
```

В принципе, все не так уж и плохо, попробуем сново отобрать признаки

```python
def get_result_LR(X_train, X_test, y_train = y_train, y_test=y_test):
    LR.fit(X_train, y_train)
    prediction_LR = LR.predict(X_test)
    return average_precision_score(y_test, prediction_LR)

X_train_LR = X_train.copy()
X_test_LR = X_test.copy()
results = []
columns_drop = []
for i in range(X_scaled_cat.shape[1]):
    results.append(get_result(X_train_LR, X_test_LR))
    X_1 = sm.add_constant(X_train_LR)
    model = sm.OLS(y_train,X_1).fit()
    if(model.pvalues.max()>0.05):
        X_train_LR.drop(columns = model.pvalues.idxmax(), axis=1, inplace=True)
        X_test_LR.drop(columns = model.pvalues.idxmax(), axis=1, inplace=True)
        columns_drop.append(model.pvalues.idxmax())
        print(i)
    else:
        break
plt.plot(columns_drop, results[:-1])
```

Качество не улучшается.Возможно наша модель переобучилась, давайте проверим это:

```python
LR.best_score_
```

Качество на обучение было дейтсвительно лучше, но не прям сильно, чтобы бить тревогу. Попробуем полиномиальное преобразование вещественных признаков, чтобы внедрить нелинейную связь

```python
from sklearn.preprocessing import PolynomialFeatures

transform = PolynomialFeatures(2)
X_train_poly = transform.fit_transform(X_train.select_dtypes(include=np.number))
X_test_poly = transform.fit_transform(X_test.select_dtypes(include=np.number))
scaler = StandardScaler()
X_train_poly = scaler.fit_transform(X_train_poly)
X_test_poly = scaler.fit_transform(X_test_poly)
```

```python
X_train_cat_poly = np.hstack((X_train.select_dtypes(include='object').values, X_train_poly))
X_test_cat_poly = np.hstack((X_test.select_dtypes(include='object').values, X_test_poly))
```

```python
%%time
LR.fit(X_train_cat_poly, y_train)
prediction_LR = LR.predict(X_test_cat_poly)
print("PR auc: ", average_precision_score(y_test, prediction_LR))
```

Точность сильно не улучшилась, мы сделали все,что было в наших силах

## Метод ближайших соседей

Т.к. время работы KNN слишком сильно возрастает от количества признаков, то просто возьмем стандартные параметры и будем надеятся на чудо, т.к. при данном количестве признаков время работы алгоритма~8 мин

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
%%time
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
prediction_KNN = knn.predict(X_test)
print("PR auc: ", average_precision_score(y_test, prediction_KNN))
```
